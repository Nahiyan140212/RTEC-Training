```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
# Histogram of patient ages
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='AgeOnIndexDate', bins=30, kde=True)
plt.title('Distribution of Patient Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

sns.set_theme(style="whitegrid")

# Count plot of patients by race
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='FirstRace',
order=df['FirstRace'].value_counts().index)
plt.title('Number of Patients by Race')
plt.xlabel('Count')
plt.ylabel('Race')
plt.show()
```
